### Now we try to clean the essay column(The meat of our dataset)

```python
from gingerit.gingerit import GingerIt
```

```python
parser=GingerIt()
```

```python
## 1. slice the string so that the parser can count the errors (slicing at full stops)
#2.sum the errors and add it to a column
#3.rectify the original column so as to make sense of the statement
```

```python
#for testing purpose
'''essays=pd.DataFrame({ 'name':['sakina','xyz','tanim','Rakesh','ramesh'],
            'age':[20,23,18,50,45],
            'essay':['Your my world .th worlt on fire','helle to the beautifullest persen in the galxy','i can do it all day.ir is a wonderful lief','the world can see not.i can her your voice','I like to play outside in the sumer adn rains but not all day']
             })
essays.head()'''
```

```python
# go to every row and select essay column

total=[]
corrected=[]
for ind in range(0,len(essays)):
    text=essays['essay'][ind]
    #we split the string on full stops assuming the sentences will not exceed the parser's max length
    strings=text.split('.')
    ans_string=[]
    tot_mist=0
    print(len(strings))
    for i in range(0,len(strings)):
        s=strings[i]
        print("The split is for ",ind,' ',s)
        mistakes=parser.parse(s)## using parse method
        tot_mist=tot_mist+len(mistakes['corrections'])#keeping a count of the number of mistakes in total
        ans_string.append(mistakes['result'])
    total.append(tot_mist)
    corrected.append(' '.join(ans_string))
    
    
        
    
essays['Total Mistakes']=total  
essays['Corrected String']=corrected
```
